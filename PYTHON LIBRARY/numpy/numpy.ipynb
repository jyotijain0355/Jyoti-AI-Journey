# TOPIC 4: Indexing & Slicing (VERY IMPORTANT + tricky)

```python
import numpy as np
arr = np.array([10, 20, 30, 40])
print(arr[0])
print(arr[1]) # 20
print(arr[2]) # 30

```

```python
#� 3. Negative Indexing �
# � Last se access karna
print(arr[-1]) # 40
print(arr[-2]) # 30
```

```python
#4. Slicing (Range Access)

print(arr[1:3])
```

```python
import numpy as np
arr = np.array([10, 20, 30, 40])
print(arr[:3]) # start se 3 tak
print(arr[2:]) # 2 se end tak
print(arr[::2])# 5. Step Slicing 

```

```python
arr = np.array([[1,2,3],[4,5,6]])
print(arr[0,1])
# #✔ row 0
# ✔ column 1

print(arr[1,2]) # 6
# 2d slicing 

print(arr[:, 1])
# Row slicing:
print(arr[0, :])
# 8. Fancy Indexing 
# Specific indexes select karna

arr = np.array([10,20,30,40])
print(arr[[0,2,3]])
```

```python
import numpy as np
arr = np.array([[1,2,3],[4,5,6]])
arr = np.array([10,20,30,40])
print(arr[arr > 20])
arr[::-1] # reverse array
```

TOPIC 8: Copy vs View (FULL DETAIL �)

```python
a = np.array([1,2,3])
b = a.copy()
b[0] = 100
print("a:", a)
print("b:", b)
```

```python
a = np.array([1,2,3])
b = a.view()
b[0] = 100
print("a:", a)
print("b:", b)
```

```python
a = np.array([1,2,3])
b = a.copy()
c = a.view()
print(b.base)
print(c.base)
```
